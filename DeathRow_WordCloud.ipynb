```
import pandas
df = pandas.read_csv('data.csv', header=0);
data = df['lastStatement']
```

```
print(data.head())
```

```
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```
lines = data.values.tolist()
line = ''.join(str(e) for e in lines)
```

```
wordcloud = WordCloud(width=1600, height=800).generate(line)

# Display the generated image:
plt.figure( figsize=(20,10), facecolor='k')
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
plt.savefig('final.png', facecolor='k', bbox_inches='tight')
```
